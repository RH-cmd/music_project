```python
import pandas as pd
```

```python
# File to Load
music_file = "static/js/data_cleaned2.csv"


# Read School and Student Data File and store into Pandas DataFrames
music_df = pd.read_csv(music_file)
```

```python
music_df
```

```python
genres_artists_df = music_df[['artist', 'genre', 'subgenre']]
genres_artists_df
```

```python
genres_list = genres_artists_df['genre'].to_list()
genres_list
```

```python
subgenres_list = genres_artists_df['subgenre'].to_list()
subgenres_list
```

```python
artists_list = genres_artists_df['artist'].to_list()
artists_list
```

```python
# Adding genre to all grandparent genres

genres_list_updated = ['genre-' + x for x in genres_list]
genres_list_updated
```

```python
genres_artists_df['genre'].unique()
```

```python
genre_parent_df = pd.DataFrame(list(zip(genres_list_updated, genres_list)), columns=['ids', 'labels'])
genre_parent_df
```

```python
genre_parent_df['parents'] = ""
genre_parent_df
```

```python
genre_parent_df.to_csv('genre_parent.csv', index = False).
```

```python
#Subgenres

genres_artists_df['genre-subgenre'] = genres_artists_df['genre'].str.cat(genres_artists_df['subgenre'], sep ="-")
genres_artists_df
```

```python
genre_sub_list = genres_artists_df['genre-subgenre'].to_list()
genre_sub_list
```

```python
subgenre_child_df = pd.DataFrame(list(zip(genre_sub_list, subgenres_list, genres_list_updated)), columns=['ids', 'labels', 'parents'])
subgenre_child_df
```

```python
subgenre_child_df.to_csv('subgenre_child.csv', index = False)
```

```python
genres_artists_df['subgenre-artist'] = genres_artists_df['subgenre'].str.cat(genres_artists_df['artist'], sep ="-")
genres_artists_df
```

```python
subgenre_artist_list_updated = genres_artists_df['subgenre-artist']
subgenre_artist_list_updated
```

```python
artist_subgenre_df = pd.DataFrame(list(zip(subgenre_artist_list_updated, artists_list, genre_sub_list)), columns=['ids', 'labels', 'parents'])
artist_subgenre_df
```

```python
all_dfs = [genre_parent_df, subgenre_child_df, artist_subgenre_df]
```

```python
for df in all_dfs:
    df.columns = ['ids', 'labels', 'parents']
```

```python
total_music_dfs = pd.concat(all_dfs).reset_index(drop=True)
```
